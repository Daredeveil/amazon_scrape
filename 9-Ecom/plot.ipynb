```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Capital Bike Share DataFrames
df = pd.read_csv('temp.csv')
```

```python
sns.set(rc={'figure.figsize':(12.7,8.27)})
fig, ax = plt.subplots()
sns.barplot(data=df[['Country','Sum_sale']], x='Country', y='Sum_sale', ax=ax, ci=None)

plt.title('Top 5 Total_Sales in Europe excluding Germany', fontsize=25)
plt.ylabel('Tolal_sale (M Euro)', fontsize=20)
plt.xlabel('Country', fontsize=20)
plt.tick_params(labelsize=15)
plt.ticklabel_format(style='plain', axis='y')
```

```python
sns.lineplot(y='Sum_sale', x='Country', data=df)
plt.ticklabel_format(style='plain', axis='y')
```

```python
sns.pointplot(data=df, x='Country', y='Sum_sale')

plt.title('', fontsize=20)
plt.ylabel('Total_Sales', fontsize=20)
plt.xlabel('Continent', fontsize=20)
plt.tick_params(labelsize=13)
plt.show()
```

```python
sns.histplot(data=df, x='Country', y='Sum_sale', multiple="dodge", shrink=1)
plt.title('', fontsize=20)
plt.xlabel('Continent', fontsize=20)
plt.ylabel('Total_Sales', fontsize=20)
plt.tick_params(labelsize=13)
plt.ticklabel_format(style='plain', axis='y')
```
